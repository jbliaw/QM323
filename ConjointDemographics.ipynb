```
import pandas as pd
```

```
filepath = '/content/QM323/Data/QM323_B3_T4_ConjointSurvey_October 5, 2020_18.46.xlsx'
df = pd.read_excel(filepath, sheet_name='Sheet1')
df.head()
```

```
df.loc[(df['AGE'] >= 18) & (df['AGE'] <= 27), 'agebuckets'] = '18-27'
df.loc[(df['AGE'] >= 28) & (df['AGE'] <= 47), 'agebuckets'] = '28-47'
df.loc[(df['AGE'] >= 48) & (df['AGE'] <= 57), 'agebuckets'] = '48-57'
df.loc[(df['AGE'] >= 58), 'agebuckets'] = '58+'
df.head()
```

```
  df = df.rename(columns={140: "$140", 
                   100: "$100", 
                   "adjustable": "Adjustable Setting",
                   "textured": "Textured Grip", 
                   "Connect longer wire": "Connected to Longer Wire"}, errors='raise')
df.head()
```

```
df1 = df.pivot_table(index='agebuckets', values='NAME', aggfunc='count')
df1
```

```
df2 = df.pivot_table(index='agebuckets', values=['$140', '$100', 'Adjustable Setting', 'Textured Grip', 'Connected to Longer Wire'])
df2
```

```
df2['$100'] = df2['$100'].round(decimals=3)
df2['$140'] = df2['$140'].round(decimals=3)
df2['Adjustable Setting'] = df2['Adjustable Setting'].round(decimals=3)
df2['Connected to Longer Wire'] = df2['Connected to Longer Wire'].round(decimals=3)
df2['Textured Grip'] = df2['Textured Grip'].round(decimals=3)

df2
```
